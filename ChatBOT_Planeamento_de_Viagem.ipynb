```python
# Instalando o SDK do Google
!pip install -q -U google-generativeai

# Importando bibliotecas necessárias
import google.generativeai as genai
from google.colab import userdata

# Configurando a API (substitua 'SUA_CHAVE_DE_API' pela sua chave)
api_key = userdata.get("SECRET_KEY")
genai.configure(api_key=api_key)

# Configurações do modelo
generation_config = {
    "candidate_count": 1,
    "temperature": 0.5,
}

safety_settings ={
    "HARASSMENT": "BLOCK_NONE",
    "HATE": "BLOCK_NONE",
    "SEXUAL": "BLOCK_NONE",
}

# Inicializando o modelo (ajuste o nome do modelo se necessário)
model = genai.GenerativeModel(model_name="gemini-1.0-pro",
                              generation_config=generation_config,
                              safety_settings=safety_settings)

# Função para obter informações da viagem
def get_trip_info():
  destino = input("Para onde você gostaria de viajar? ")
  data_inicio = input("Qual é a data de início da viagem? ")
  data_fim = input("Qual é a data de fim da viagem? ")
  orcamento = input("Qual é o seu orçamento aproximado? ")
  interesses = input("Quais são os seus principais interesses (ex: história, natureza, gastronomia)? ")
  return destino, data_inicio, data_fim, orcamento, interesses

# Função para gerar sugestões de viagem
def gerar_sugestoes(destino, data_inicio, data_fim, orcamento, interesses):
  prompt = f"Sugestões de viagem para {destino} de {data_inicio} a {data_fim} com orçamento de {orcamento} e interesse em {interesses}"
  response = model.generate_content(prompt)
  return response.text

# Início da interação com o chatbot
print("Olá! Sou o seu chatbot de planeamento de viagens.")

# Loop principal
while True:
  opcao = input("O que você gostaria de fazer? (1 - Planear viagem, 2 - Sair) ")
  if opcao == "1":
    destino, data_inicio, data_fim, orcamento, interesses = get_trip_info()
    sugestoes = gerar_sugestoes(destino, data_inicio, data_fim, orcamento, interesses)
    print("Aqui estão algumas sugestões para sua viagem:\n", sugestoes)
  elif opcao == "2":
    print("Obrigado por usar o chatbot de planeamento de viagens!")
    break
  else:
    print("Opção inválida. Por favor, escolha 1 ou 2.")
```
